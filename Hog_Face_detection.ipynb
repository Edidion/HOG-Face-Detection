```python
#import libraries
import dlib
import cv2
```

```python
#read image from image path in drive
image = cv2.imread('/content/drive/MyDrive/Comp Vision/Images-20210601T134554Z-001.zip (Unzipped Files)/Images/people2.jpg')
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

```python
#create histogram of oriented gradients face detector 
hog_face_detector =dlib.get_frontal_face_detector()
```

```python
#create detections variable. The image scale is set as 1
detections = hog_face_detector(image, 1)
```

```python
detections, len(detections)
```

```python
#build a for loop to detect faces in image
for face in detections:
  #print(face)
  #print(face.left())
  #print(face.top())
  #print(face.right())
  #print(face.bottom())
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l, t), (r, b), (0, 255, 255), 2)
cv2_imshow(image)
```
